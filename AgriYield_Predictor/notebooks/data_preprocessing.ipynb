```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
yield_path = "../data/raw/Yield_dataset.csv"
yield_df = pd.read_csv(yield_path)
print("YIELD DATASET:")
print(yield_df.head(), "\n")
```

```python
soil_path  = "../data/raw/Soil_dataset.csv"
soil_df  = pd.read_csv(soil_path)
print("SOIL DATASET:")
print(soil_df.head())
```

```python
yield_india_df = yield_df[yield_df["Area"] == "India"].copy()

print("Yield (India only) head:")
print(yield_india_df.head())
print("Yield (India only) shape:", yield_india_df.shape)
```

```python
#Data Normalization so the multiple values in the same feature can be separated and can be presented as single
#value in single attribute to  increase  the prediction accuracy and quality
print("Before explosion:",yield_india_df.head())
print("Shape before explosion:", yield_india_df.shape)


yield_india_df["Item"] = yield_india_df["Item"].astype(str).str.split(",")


yield_exploded_df = yield_india_df.explode("Item").reset_index(drop=True)


yield_exploded_df["Item"] = yield_exploded_df["Item"].str.strip().str.lower()

print("\nAfter explosion:")
print(yield_exploded_df.head())
print("Shape after explosion:", yield_exploded_df.shape)
```
