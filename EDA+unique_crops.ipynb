```python
# Mount Google Drive and import datasets
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np

base_path = "/content/drive/MyDrive/AgriYield_Data/"

# Load datasets
fao = pd.read_csv(base_path + "FAOSTAT_data_en_11-19-2025.csv")
crop = pd.read_csv(base_path + "Crop_recommendation.csv")

print("FAOSTAT shape:", fao.shape)
print("Crop recommendation shape:", crop.shape)

# Display first few rows of each dataset
print("\nFAOSTAT Data Preview:")
display(fao.head())

print("\nCrop Recommendation Data Preview:")
display(crop.head())
```

```python
# Display the full FAOSTAT dataframe
display(fao)
```

```python
# Display the full crop dataframe
display(crop)
```

```python
# Show unique crop items in FAOSTAT data (first 40)
print("Unique crop items in FAOSTAT data (first 40):")
display(fao['Item'].unique()[:40])
```

```python
# Print all unique crop items in FAOSTAT data
print("All unique crop items in FAOSTAT data:")
for i in fao['Item'].unique():
    print(i)
```

```python
# Show unique crop labels in crop recommendation data
print("Unique crop labels in crop recommendation data:")
display(crop['label'].unique())
```

```python
# Split FAOSTAT items by commas or "and" and explode into separate rows
fao['Item'] = fao['Item'].str.split(r",|and")
fao = fao.explode('Item')
fao['Item'] = fao['Item'].str.strip()

# Display the transformed FAOSTAT data
print("FAOSTAT data after splitting items:")
display(fao)
```

```python
# Clean up item names
fao['Item'] = fao['Item'].str.replace('berries', '').str.replace('raw', '').str.replace('n.e.c.', '').str.replace('excluding dehydrated', '').str.strip()
fao['Item'] = fao['Item'].str.replace('\\(excluding shelled\\)', '', regex=True).str.strip()
fao['Item'] = fao['Item'].str.replace('\\(Capsicum spp\\. and Pimenta spp\\.\\)', '', regex=True).str.strip()
fao['Item'] = fao['Item'].str.replace('\\(Capsicum spp\\.\\, Pimenta spp\\.\\)', '', regex=True).str.strip()

# Remove extra spaces and special characters
fao['Item'] = fao['Item'].str.replace('\\.', '', regex=True)
fao['Item'] = fao['Item'].str.strip()

# Show unique items after cleaning
print("Unique items after cleaning:")
display(fao['Item'].unique()[:50])
```

```python
# Define crop mapping dictionary
crop_mapping = {
    "Rice": "rice",
    "Maize corn": "maize",
    "Maize": "maize",
    "Chick peas": "chickpea",
    "Pigeon peas": "pigeonpeas",
    "Beans dry": "mothbeans",
    "Beans": "mothbeans",
    "Lentils": "lentil",
    "Bananas": "banana",
    "Mangoes guavas and mangosteens": "mango",
    "Mangoes": "mango",
    "Grapes": "grapes",
    "Watermelons": "watermelon",
    "Cantaloupes": "muskmelon",
    "Cantaloupes and other melons": "muskmelon",
    "Apples": "apple",
    "Oranges": "orange",
    "Papayas": "papaya",
    "Coconuts": "coconut",
    "Seed cotton unginned": "cotton",
    "Seed cotton": "cotton",
    "Jute": "jute",
    "Coffee green": "coffee",
    "Coffee": "coffee"
}

print("Number of crop mappings:", len(crop_mapping))
```

```python
# Map crop names in FAOSTAT data
fao['Crop'] = fao['Item'].map(crop_mapping)

# Check mapping results
print("Number of rows with successful mapping:", fao['Crop'].notna().sum())
print("Number of rows without mapping:", fao['Crop'].isna().sum())

# Show crops that didn't get mapped
print("\nItems that didn't get mapped:")
unmapped_items = fao[fao['Crop'].isna()]['Item'].unique()
for item in unmapped_items[:20]:  # Show first 20
    print(item)
```

```python
# Rename 'label' column in crop dataframe to 'Crop' for merging
crop = crop.rename(columns={'label': 'Crop'})

# Display the renamed crop dataframe
print("Crop dataframe after renaming:")
display(crop.head())
```

```python
# Merge the two datasets on 'Crop' column
df_merged = crop.merge(
    fao,
    left_on="Crop",
    right_on="Crop",
    how="inner"
)

print("Merged dataframe shape:", df_merged.shape)
display(df_merged.head())
```

```python
# Filter only 'Production' rows from the merged data
df_merged = df_merged[df_merged['Element'] == 'Production']

print("Dataframe after filtering for Production only:")
print("Shape:", df_merged.shape)
display(df_merged.head())
```

```python
# Check distribution of crops in the merged data
print("Crop distribution in merged data:")
crop_counts = df_merged['Crop'].value_counts()
display(crop_counts)
```

```python
# Drop unnecessary columns
columns_to_drop = ['Domain Code', 'Domain', 'Area Code (M49)', 'Area', 'Element Code',
                  'Element', 'Item Code (CPC)', 'Item', 'Year', 'Year Code',
                  'Unit', 'Flag', 'Flag Description', 'Note']

df_merged = df_merged.drop(columns=columns_to_drop)

print("Dataframe after dropping unnecessary columns:")
print("Shape:", df_merged.shape)
display(df_merged.head())
```

```python
# Check for null values
print("Null values in each column:")
display(df_merged.isnull().sum())
```

```python
# Check for duplicate rows
duplicate_count = df_merged.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Remove duplicates and reset index
df_merged = df_merged.drop_duplicates().reset_index(drop=True)
print(f"Dataframe shape after removing duplicates: {df_merged.shape}")
display(df_merged.head())
```

```python
# Generate descriptive statistics
print("Descriptive statistics:")
display(df_merged.describe())
```

```python
# Check number of unique values in each column
print("Number of unique values in each column:")
display(df_merged.nunique())
```

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [10, 6]

# Define numeric columns
numeric_cols = ["N", "P", "K", "temperature", "humidity", "ph", "rainfall", "Value"]
```

```python
# Remove outliers using IQR method
Q1 = df_merged[numeric_cols].quantile(0.25)
Q3 = df_merged[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Create a mask for rows without outliers
mask = ~((df_merged[numeric_cols] < (Q1 - 1.5 * IQR)) |
         (df_merged[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

final_clean = df_merged[mask]

print(f"Before outlier removal: {df_merged.shape}")
print(f"After outlier removal: {final_clean.shape}")
print(f"Rows removed: {df_merged.shape[0] - final_clean.shape[0]}")
```

```python
# Display cleaned data
print("Cleaned data preview:")
display(final_clean.head())
```

```python
print(f"Total rows in final cleaned data: {len(final_clean)}")
```

```python
# Check crop distribution in cleaned data
print("Crop distribution in cleaned data:")
clean_crop_counts = final_clean['Crop'].value_counts()
display(clean_crop_counts)
```

```python
# Compare crops before and after cleaning
all_crops = set(df_merged['Crop'].unique())
clean_crops = set(final_clean['Crop'].unique())

print("Crops Before Cleaning:", len(all_crops), sorted(all_crops))
print("Crops After Cleaning:", len(clean_crops), sorted(clean_crops))

removed_crops = sorted(all_crops - clean_crops)
print("\nRemoved Crops:", removed_crops)
```

```python
# Create boxplots to visualize the distribution after outlier removal
plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(y=final_clean[col])
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.suptitle('Distribution of Numerical Features After Outlier Removal', y=1.02, fontsize=16)
plt.show()
```

```python
# Create a single boxplot for all numeric features (like in the original notebook)
plt.figure(figsize=(12, 5))
sns.boxplot(data=final_clean[numeric_cols])
plt.title("Boxplot after removing outliers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Additional visualizations: Histograms for numeric features
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(final_clean[col], bins=30, edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

plt.suptitle('Histograms of Numerical Features', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = final_clean[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```
