### 1. Boxplots BEFORE Outlier Removal

### 2. Remove Outliers (IQR Method)

```python
import numpy as np

def cap_outliers(df, cols):
    df_cap = df.copy()
    for col in cols:
        Q1 = df_cap[col].quantile(0.25)
        Q3 = df_cap[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df_cap[col] = np.where(df_cap[col] < lower, lower, df_cap[col])
        df_cap[col] = np.where(df_cap[col] > upper, upper, df_cap[col])
    return df_cap

# Define the list of numeric columns
numeric_cols = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall', 'Value']

cleaned = cap_outliers(final_data, numeric_cols)

print("Shape Before:", final_data.shape)
print("Shape After:", cleaned.shape)
```

### 3. Boxplots AFTER Outlier Removal

##  Encode Categorical Feature (Label Encoding)

### 1. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cleaned['label_encoded'] = le.fit_transform(cleaned['label'])
# Crop type is a categorical independent variable,
# encoded numerically to be used in regression.
```

```python
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label Encoding Mapping:")
for crop, code in label_mapping.items():
    print(f"{crop} → {code}")

cleaned[['label', 'label_encoded']]
```

### 2. Feature Scaling
(Applied only on numerical features)

```python
# Scaling converts all numeric values to the same scale so no feature dominates the model.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = ['N','P','K','temperature','humidity','ph','rainfall']


cleaned[num_cols] = scaler.fit_transform(cleaned[num_cols])
cleaned[num_cols]
```

### 3. Feature Engineering


---

You added two new scientific features:

1. Season Index

   * This represents how climate favors growth.

   * Higher index → better growing conditions.
2. Water Stress

   * This indicates whether a crop gets enough water relative to heat.

   * Low water stress → dry conditions
   * High water stress → wet conditions




```python
# Growing season index
# Based on temperature + humidity

cleaned['season_index'] = cleaned['temperature'] * cleaned['humidity']
cleaned['season_index']
```

```python
# Water stress index

cleaned['water_stress'] = cleaned['rainfall'] / cleaned['temperature']
cleaned['water_stress']
```
