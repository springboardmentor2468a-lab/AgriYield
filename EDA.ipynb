```python
import pandas as pd
import re

# Read datasets
df_crop = pd.read_csv("Crop_recommendation.csv")
df_faostat = pd.read_csv("Exploded_FAOSTAT_data.csv")
```

```python
def normalize_name(name):
    name = str(name).lower().strip()

    # Common plural/suffix patterns handled explicitly
    replacements = {
        'apples': 'apple',
        'mangoes': 'mango',
        'berries': 'berry',
        'grapes': 'grape',
        'beans': 'bean',
        'peas': 'pea',
        'oranges': 'orange',
        'melons': 'melon',
        'bananas': 'banana',
        'pomegranates': 'pomegranate',
        'tomatoes': 'tomato',
        'potatoes': 'potato',
        'chickpeas': 'chickpea',
        'kidneybeans': 'kidneybean',
        'pigeonpeas': 'pigeonpea',
        'mungbeans': 'mungbean',
        'mothbeans': 'mothbean',
        'blackgrams': 'blackgram',
        'lentils': 'lentil',
        'cottons': 'cotton'
    }

    # Replace exact plural forms first
    for plural, singular in replacements.items():
        if name.endswith(plural):
            return singular

    # Generic rules for remaining plural endings
    name = re.sub(r'ies$', 'y', name)
    name = re.sub(r'oes$', 'o', name)
    name = re.sub(r'es$', '', name)
    name = re.sub(r's$', '', name)

    return name
```

```python
# Normalize label columns in both datasets
df_crop["norm_label"] = df_crop["label"].apply(normalize_name)
df_faostat["norm_label"] = df_faostat["label"].apply(normalize_name)
```

```python
merged_df = pd.merge(df_crop, df_faostat, on="norm_label", how="outer", suffixes=("_crop", "_faostat"))
```

```python
# Drop duplicates if any
merged_df = merged_df.drop_duplicates()

# Save the merged dataset
merged_df.to_csv("Merged_Crop_FAOSTAT.csv", index=False)

print("âœ… Merged dataset saved as: Merged_Crop_FAOSTAT.csv")
print("Total rows in merged dataset:", len(merged_df))
```

```python
# Which labels in crop dataset didnâ€™t find match in FAOSTAT
unmatched_crop = set(df_crop["norm_label"]) - set(df_faostat["norm_label"])
print("ðŸš« Unmatched crop labels:", unmatched_crop)

# Which FAOSTAT labels didnâ€™t match crop dataset
unmatched_faostat = set(df_faostat["norm_label"]) - set(df_crop["norm_label"])
print("ðŸš« Unmatched FAOSTAT labels:", list(unmatched_faostat)[:20])  # first 20 for view
```

```python
import pandas as pd

# Read the filtered merged dataset
df = pd.read_csv("Merged_Crop_FAOSTAT_Filtered.csv")
```

```python
# Columns to remove
cols_to_drop = [
    'Domain Code', 'Domain', 'Area Code (M49)', 'Area',
    'Element Code', 'Item Code (CPC)', 'Year Code', 'Year',
    'Unit', 'Flag', 'Flag Description', 'Note'
]

# Drop only if they exist
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])
```

```python
# Combine label_crop, norm_label, and label_faostat into a single 'label' column
df["label"] = df["label_crop"].combine_first(df["norm_label"]).combine_first(df["label_faostat"])

# Drop the old label columns
df = df.drop(columns=["label_crop", "norm_label", "label_faostat"], errors='ignore')
```

```python
# Drop rows with nulls in key numeric columns
df = df.dropna(subset=["N", "P", "K", "temperature", "humidity", "ph", "rainfall", "Value"], how="any")
```

```python
# Columns to check for outliers
numeric_cols = ["N", "P", "K", "temperature", "humidity", "ph", "rainfall", "Value"]

# Remove outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# Save to new file
df.to_csv("Cleaned_Crop_FAOSTAT.csv", index=False)

print("âœ… Cleaned dataset saved as: Cleaned_Crop_FAOSTAT.csv")
print("Rows remaining:", len(df))
```
