```python
df1['Item']  = df1['Item'].str.strip().str.lower()
df2['label'] = df2['label'].str.strip().str.lower()

merged_df = pd.merge(
    df1,
    df2,
    how='inner',
    left_on='Item',
    right_on='label'
)

merged_df.drop(columns=['label'], inplace=True)
display(merged_df.head())
print("Merged shape:", merged_df.shape)
merged_df.to_excel("merged_df.xlsx", index=False)
```

```python
merged_elements = merged_df['Element'].unique()
# display(merged_elements)
final_df = merged_df[merged_df['Element'] == 'Production'].copy()
final_elements = final_df['Element'].unique()
display(final_elements)
```

```python
final_df = final_df.drop(columns=['Element'])
display(final_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1. Print Null Values
# -----------------------------
print("Null values in each column:")
print(final_df.isnull().sum())

# -----------------------------
# 2. Drop / Impute Nulls
# -----------------------------
# Drop rows with nulls
final_df = final_df.dropna()

# Or impute numeric columns with median
# for col in final_df.select_dtypes(include=np.number).columns:
#     final_df[col].fillna(final_df[col].median(), inplace=True)

# -----------------------------
# 3. Print Duplicates
# -----------------------------
duplicates = final_df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")

# -----------------------------
# 4. Drop Duplicates
# -----------------------------
final_df = final_df.drop_duplicates()

# -----------------------------
# 5. Removing Outliers (IQR method)
# -----------------------------
numeric_cols = final_df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    Q1 = final_df[col].quantile(0.25)
    Q3 = final_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    final_df = final_df[(final_df[col] >= lower) & (final_df[col] <= upper)]

# -----------------------------
# 6. Save Clean Dataset
# -----------------------------
final_df.to_csv('cleaned_dataset.csv', index=False)
print("Clean dataset saved as 'cleaned_dataset.csv'")

# -----------------------------
# 7. Generate Correlation Heatmap (matplotlib only)
# -----------------------------
corr = final_df.corr()
plt.figure(figsize=(10, 8))
plt.imshow(corr, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Heatmap (Matplotlib)")
plt.show()

# -----------------------------
# 8. Generate Bar Chart (matplotlib only)
# -----------------------------
categorical_col = 'Country' if 'Country' in final_df.columns else final_df.select_dtypes(include='object').columns[0]

counts = final_df[categorical_col].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(counts.index, counts.values)
plt.title(f"Bar Chart of {categorical_col}")
plt.xlabel(categorical_col)
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
