```python
numeric_cols = df3.columns
outlier_summary = {}
for col in numeric_cols:
    Q1 = df3[col].quantile(0.25)
    Q3 = df3[col].quantile(0.85)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df3[(df3[col] < lower_bound) | (df3[col] > upper_bound)][col]

    outlier_summary[col] = {
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": outliers.shape[0],
        "Outlier Values": outliers.values
    }
for col, info in outlier_summary.items():
    print("\nColumn:", col)
    print("Lower Bound:", info["Lower Bound"])
    print("Upper Bound:", info["Upper Bound"])
    print("Outliers:", info["Outlier Values"])
    print("count:",len(info['Outlier Values']))
```
