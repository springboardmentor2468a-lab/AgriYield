```python
df2["Area_Harvested"] = df2.apply(
    lambda row: df2[(df2["Item"] == row["Item"]) &
                    (df2["Year"] == row["Year"]) &
                    (df2["Element"] == "Area harvested")]["Value"].max(),
    axis=1
)
```

```python
df2["Production"] = df2.apply(
    lambda row: df2[(df2["Item"] == row["Item"]) &
                    (df2["Year"] == row["Year"]) &
                    (df2["Element"] == "Production")]["Value"].max(),
    axis=1
)
```

```python
df2["Yield"] = df2.apply(
    lambda row: df2[(df2["Item"] == row["Item"]) &
                    (df2["Year"] == row["Year"]) &
                    (df2["Element"] == "Yield")]["Value"].max(),
    axis=1
)
```
