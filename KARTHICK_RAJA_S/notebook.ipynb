```python
numeric_cols = df4.select_dtypes(include=[np.number]).columns
outlier_summary = {}
for col in numeric_cols:
    Q1 = df4[col].quantile(0.25)
    Q3 = df4[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df4[(df4[col] < lower_bound) | (df4[col] > upper_bound)][col]
    outlier_summary[col] = {
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": outliers.shape[0],
        "Outlier Values": outliers.values
    }
for col, info in outlier_summary.items():
    print("\nColumn:", col)
    print("Lower Bound:", info["Lower Bound"])
    print("Upper Bound:", info["Upper Bound"])
    print("Outliers:", info["Outlier Values"])
    print("Count:", info["Outlier Count"])
```

```python
df4_clean = df4.copy()

for col in numeric_cols:
    Q1 = df4_clean[col].quantile(0.25)
    Q3 = df4_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df4_clean = df4_clean[(df4_clean[col] >= lower_bound) & (df4_clean[col] <= upper_bound)]
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
encoder=LabelEncoder()
```

```python
df4_clean['Crops']=encoder.fit_transform(df4_clean['Crop'])
```

```python
df4_clean.head()
```

```python
df4_clean=df4_clean.drop(columns=['Crop'])
```

```python
df4_clean.head()
```

```python
X=df4_clean.drop(columns=['Value'])
y=df4_clean['Value']
```

```python
X.head()
```

```python
y
```

```python
scaler=StandardScaler()
```

```python
X_scaler=scaler.fit_transform(X)
X_scaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaler,y,random_state=32,test_size=0.25)
X_train,X_test,y_train,y_test
```

```python
from xgboost import XGBRegressor
```

```python
xg_model=XGBRegressor()
```

```python
xg_model.fit(X_train,y_train)
```

```python
xg_model.score(X_test,y_test)
y_pred=xg_model.predict(X_test)
```

```python
y_train_pred=xg_model.predict(X_train)
```

```python
mse=mean_squared_error(y_pred=y_train_pred,y_true=y_train)
mae=mean_absolute_error(y_pred=y_train_pred,y_true=y_train)
rmse=root_mean_squared_error(y_pred=y_train_pred,y_true=y_train)
r2=r2_score(y_pred=y_train_pred,y_true=y_train)
print("train Mean square error=",mse)
print("train Mean abosule error=",mae)
print("train Root mean square error=",rmse)
print("train R2 score=",r2)
```

```python
mse=mean_squared_error(y_pred=y_pred,y_true=y_test)
mae=mean_absolute_error(y_pred=y_pred,y_true=y_test)
rmse=root_mean_squared_error(y_pred=y_pred,y_true=y_test)
r2=r2_score(y_pred=y_pred,y_true=y_test)
print("test Mean square error=",mse)
print("test Mean abosule error=",mae)
print("test Root mean square error=",rmse)
print("test R2 score=",r2)
```
